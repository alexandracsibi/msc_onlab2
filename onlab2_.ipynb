If you drew triangles between all spatial points such that no triangle contains another point inside it, that's a Delaunay triangulation. Spots connected by triangle edges become neighbors.

Use when:
 - You want a more natural, geometry-based graph (not forced by k).
 - Your data is on a non-grid layout (Slide-seq, Stereo-seq, custom).
 - You want a sparser and topology-respecting graph.

```python
# Delaunay triangulation graph
sq.gr.spatial_neighbors(adata, coord_type='generic', delaunay=True)

sq.pl.spatial_scatter(adata, connectivity_key='spatial_connectivities', size=1.5)
```

```python
# Extract expression to tensor
x = torch.tensor(adata.X, dtype=torch.float)

# Spatial graph into edge index
coord = sp.coo_matrix(adata.obsp['spatial_connectivities'])
edge_index = torch.tensor(np.vstack((coord.row, coord.col)), dtype=torch.long)

# PyTorch Geometric graph data
graph_data = tg.data.Data(x=x, edge_index=edge_index)
graph_data
```

## VGAE implementation

```python
import torch.nn.functional as F
# Encoder and decoder
class Encoder(torch.nn.Module):
    def __init__(self, in_channels, out_channels):
        super(Encoder, self).__init__()
        self.conv1 = tg.nn.GCNConv(in_channels, 128)
        self.conv_mu = tg.nn.GCNConv(128, out_channels)
        self.conv_logstd = tg.nn.GCNConv(128, out_channels)

    def forward(self, x, edge_index):
        x = F.relu(self.conv1(x, edge_index))
        return self.conv_mu(x, edge_index), self.conv_logstd(x, edge_index)

# Instantiate VGAE model
out_channels = 16  # embedding size
model = tg.nn.VGAE(Encoder(graph_data.num_features, out_channels))
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

## Train VGAE

```python
model.train()
epochs = 100

for epoch in range(epochs):
    optimizer.zero_grad()
    z = model.encode(graph_data.x, graph_data.edge_index)
    loss = model.recon_loss(z, graph_data.edge_index) + (1 / graph_data.num_nodes) * model.kl_loss()
    loss.backward()
    optimizer.step()

    if epoch % 10 == 0:
        print(f'Epoch: {epoch}, Loss: {loss.item():.4f}')
```

## Generate node (vertex) embeddings

```python
model.eval()
with torch.no_grad():
    embeddings = model.encode(graph_data.x, graph_data.edge_index).cpu().numpy()

print("Embeddings shape:", embeddings.shape)
```

## Clustering embeddings

```python
# Determine optimal number of clusters (example: 5 clusters)
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
clusters = kmeans.fit_predict(embeddings)

# Evaluate clustering quality (Silhouette score)
sil_score = silhouette_score(embeddings, clusters)
print(f'Silhouette Score: {sil_score:.4f}')
```

```python
# Visualization
adata.obs['GNN_clusters'] = clusters.astype(str)
sq.pl.spatial_scatter(adata, color='GNN_clusters', title="Clusters from GNN embeddings")
```

##  Decoder for denoising (reconstruction)

```python
model.eval()
with torch.no_grad():
    z = model.encode(graph_data.x, graph_data.edge_index)
    reconstructed_adj = model.decoder.forward_all(z)

# Original adjacency matrix
original_adj = torch.zeros((graph_data.num_nodes, graph_data.num_nodes))
original_adj[graph_data.edge_index[0], graph_data.edge_index[1]] = 1

# Calculate reconstruction accuracy (AUC)
from sklearn.metrics import roc_auc_score

# Flatten matrices for AUC computation
original_flat = original_adj.cpu().numpy().flatten()
reconstructed_flat = reconstructed_adj.cpu().numpy().flatten()

auc_score = roc_auc_score(original_flat, reconstructed_flat)
print(f"Reconstruction AUC Score (Denoising Performance): {auc_score:.4f}")
```
